```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/Position_Salaries.csv')
df
```

```python
features = df.iloc[:, 1:-1].values
labels = df.iloc[:, -1].values
```

```python
print(features)
print(labels)
```

```python
plt.scatter(features, labels)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(features, labels)
```

```python
plt.scatter(features, labels)
plt.plot(features, lr.predict(features))
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures

def degree():
  for x in range(2, 10):
    polynomial_reg = PolynomialFeatures(degree = x)
    features_poly = polynomial_reg.fit_transform(features)
    lr2 = LinearRegression()
    lr2.fit(features_poly, labels)
    print("Degree: ",x)
    print("Accuracy: ",lr2.score(features_poly, labels))

degree()
```

```python
polynomial_reg = PolynomialFeatures(degree = 3)
features_poly = polynomial_reg.fit_transform(features)
lr2 = LinearRegression()
lr2.fit(features_poly, labels)
```

```python
plt.scatter(features, labels)
plt.plot(features, lr2.predict(features_poly))
plt.show()
```
