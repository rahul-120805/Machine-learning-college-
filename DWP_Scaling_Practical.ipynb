Normalisation(Min-max scalar): It is a scaling technique in which values ae shifted and re-scaled ranging between 0 and 1.
normalisation.png


Standardization: Another scaling technique where values are centred arround mean with a unit S.D.


standardization.png

When data is not normally distributed, we use normalisation; if data is normally distributed, we use standardization.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/refs/heads/master/BostonHousing.csv")
df.info()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)
normalized_data
```

```python
new_df = pd.DataFrame(normalized_data, columns=df.columns)
new_df
```

```python
new_df.describe()
```

```python
scaler2 = StandardScaler()
standardized_data = scaler2.fit_transform(df)
standardized_data
```

```python
new_df2 = pd.DataFrame(standardized_data, columns=df.columns)
new_df2
```

```python
from sklearn.preprocessing import RobustScaler

# Create a RobustScaler object
rb = RobustScaler()

# Fit the scaler to your data and transform it
robust_data = rb.fit_transform(df)
```

```python
new_data_df_3 = pd.DataFrame(robust_data, columns=df.columns)
new_data_df_3
```
